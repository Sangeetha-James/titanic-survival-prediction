```python
x=train_titanic.drop("Survived",axis=1)
y=train_titanic["Survived"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
train_data_predict=model.predict(x_train)
train_data_accuracy=accuracy_score(y_train,train_data_predict)
print("The accuracy of training data is",train_data_accuracy)
```

```python
test_data_predict=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,test_data_predict)
print("The accuracy of testing data is",test_data_accuracy)
```

```python
cm=confusion_matrix(test_data_predict,y_test)
cm
```

```python
sns.heatmap(cm,annot=True,annot_kws={"size":20},cmap="viridis",fmt="0.2f",square=True,cbar=True,xticklabels=['Predicted_True','Predicted False'],yticklabels=['Actual_True','Actual_False'])
plt.xlabel("Predicted Data")
plt.ylabel("Actual Data")
plt.title("Confusion matrix")
plt.show()
```

```python
print(classification_report(test_data_predict,y_test))
```

```python
new_data=(	3,	1,	26.000000	,0	,0	,7.9250,	0)
new_data_as_np=np.asarray(new_data)
new_data_reshaped=new_data_as_np.reshape(1,-1)
prediction=model.predict(new_data_reshaped)
if prediction[0]==0:
  print("The person could not survive")
else:
  print("The person survived")
```

```python
proba = model.predict_proba(new_data_reshaped)
print(proba)
```

```python
not_survived_prob = proba[0][0]
survived_prob = proba[0][1]
```

```python
print(f"Not Survived: {not_survived_prob*100:.2f}%")
print(f"Survived: {survived_prob*100:.2f}%")
```

```python
if survived_prob > 0.5:
    print(f"The person is survived ({survived_prob*100:.2f}% confidence)")
else:
    print(f"The person is not survived ({not_survived_prob*100:.2f}% confidence)")
```
